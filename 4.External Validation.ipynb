```python
import pandas as pd
df = pd.read_parquet(r'E:\DataBase\新疆数据\His\新疆住院病案首页.parquet')
```

```python
df.rename(columns={'总费用':'住院费'}, inplace=True)
```

```python
## 医疗付款方式
payment_map = {
    1.1: 'UEBMI',   # 城镇职工基本医疗保险
    2.1: 'URRBMI',   # 城镇居民基本医疗保险
    3.1: 'URRBMI',   # 新型农村合作医疗
    4.0: 'GS',   # 贫困救助
    5.0: 'COI',   # 商业医疗保险
    6.0: 'GS',   # 全公费
    7.0: 'SP',   # 全自费
    8.0: 'COI',   # 其他社会保险
    9.0: 'COI',   # 其他
}

df['医疗付款方式'] = df['医疗付款方式'].map(payment_map)
```

```python
## 入院时间
df['入院时间'] = pd.to_datetime(df['入院时间'].str[:10], errors='coerce')
df = df[(2019 <= df['入院时间'].dt.year) & (2024 >= df['入院时间'].dt.year)].copy()
print('住院时间在19-24年有', df.shape[0], '条住院记录')
```

```python
## 年龄
df['年龄'].describe()
```

```python
## 性别
df['性别']

sex_map = {
    1: 1,
    2: 0,
    9: pd.NA,
}

df['性别'] = df['性别'].map(sex_map)
```

```python
## 身份证号
import numpy as np
import re
# 身份证号规范化
pattern = r'^[1-9]\d{5}(?:18|19|20)\d{2}(?:0[1-9]|1[0-2])(?:0[1-9]|[12]\d|3[01])\d{3}[\dXx]$'

def check_id_format(s):
    if pd.isna(s):
        return np.nan
    if bool(re.match(pattern, s)):
        return s
    else:
        return np.nan
df['身份证号'] = df['身份证号'].map(check_id_format)
```

```python
## icd
icdcols = [
        '出院主要诊断编码', '出院其他诊断编码1', '出院其他诊断编码2', '出院其他诊断编码3', '出院其他诊断编码4', '出院其他诊断编码5', 
        '出院其他诊断编码6', '出院其他诊断编码7', '出院其他诊断编码8', '出院其他诊断编码9', '出院其他诊断编码10', '出院其他诊断编码11',
        '出院其他诊断编码12', '出院其他诊断编码13', '出院其他诊断编码14', '出院其他诊断编码15', '出院其他诊断编码16', '出院其他诊断编码17',
        '出院其他诊断编码18', '出院其他诊断编码19', '出院其他诊断编码20', '出院其他诊断编码21', '出院其他诊断编码22', '出院其他诊断编码23',
        '出院其他诊断编码24', '出院其他诊断编码25', '出院其他诊断编码26', '出院其他诊断编码27', '出院其他诊断编码28', '出院其他诊断编码29',
        '出院其他诊断编码30', '出院其他诊断编码31', '出院其他诊断编码32', '出院其他诊断编码33', '出院其他诊断编码34', '出院其他诊断编码35',
        '出院其他诊断编码36', '出院其他诊断编码37', '出院其他诊断编码38', '出院其他诊断编码39', '出院其他诊断编码40',
]
icd_pattern = re.compile(r'^[A-Za-z]\d{2}$')

def normalize_icd_precise(code):
    
    # 如果是空值，返回空值
    if pd.isna(code):
        return np.nan
    
    # 保留到小数点后一位
    if "." in code:
        prefix, _ = code.split(".", 1)
    
    else:
        return np.nan

    # 用正则判断是否合法ICD编码
    if icd_pattern.match(prefix):
        return prefix.upper()
    else:
        return np.nan

df[icdcols] = df[icdcols].astype(str).map(normalize_icd_precise)
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# ========================
# 1. 读取数据
# ========================
df = pd.read_parquet(r"File/hpconcat_external_validation.parquet").copy()
df = df.reset_index(drop=True)

df["入院时间"] = pd.to_datetime(df["入院时间"], errors="coerce")
df = df.dropna(subset=["身份证号", "入院时间"]).copy()

# ========================
# 2. ICD 编码定义
# ========================
icd_cols = [
    '出院主要诊断编码', '出院其他诊断编码1', '出院其他诊断编码2', '出院其他诊断编码3', '出院其他诊断编码4', '出院其他诊断编码5', 
    '出院其他诊断编码6', '出院其他诊断编码7', '出院其他诊断编码8', '出院其他诊断编码9', '出院其他诊断编码10', '出院其他诊断编码11',
    '出院其他诊断编码12', '出院其他诊断编码13', '出院其他诊断编码14', '出院其他诊断编码15', '出院其他诊断编码16', '出院其他诊断编码17',
    '出院其他诊断编码18', '出院其他诊断编码19', '出院其他诊断编码20', '出院其他诊断编码21', '出院其他诊断编码22', '出院其他诊断编码23',
    '出院其他诊断编码24', '出院其他诊断编码25', '出院其他诊断编码26', '出院其他诊断编码27', '出院其他诊断编码28', '出院其他诊断编码29',
    '出院其他诊断编码30', '出院其他诊断编码31', '出院其他诊断编码32', '出院其他诊断编码33', '出院其他诊断编码34', '出院其他诊断编码35',
    '出院其他诊断编码36', '出院其他诊断编码37', '出院其他诊断编码38', '出院其他诊断编码39', '出院其他诊断编码40',
]

heart_codes = [
    "I20", "I21", "I22", "I23", "I24", "I25",
    "I26", "I27", "I28",
    "I30", "I31", "I32", "I33", "I34", "I35", "I36", "I37", "I38", "I39",
    "I40", "I41", "I42", "I43", "I44", "I45", "I46", "I47", "I48", "I49",
    "I50", "I51", "I52"
]
brain_codes = ["I60", "I61", "I62", "I63", "I64", "I65", "I66", "I67", "I68", "I69"]
kidney_codes = ["N04", "N18"]

heart_set = set(heart_codes)
brain_set = set(brain_codes)
kidney_set = set(kidney_codes)
```

```python
# ========================
# 3. 排序并提取首次住院
# ========================
df = df.sort_values(["身份证号", "入院时间"]).copy()
df_first = df.groupby("身份证号", as_index=False).first().copy()

# ========================
# 4. 主分析：首次住院必须为 state1
#    基线用全部诊断列判断是否已有目标并发症
# ========================
first_heart = df_first[icd_cols].isin(heart_set).any(axis=1)
first_brain = df_first[icd_cols].isin(brain_set).any(axis=1)
first_kidney = df_first[icd_cols].isin(kidney_set).any(axis=1)

bad_ids = df_first.loc[first_heart | first_brain | first_kidney, "身份证号"].unique()
df_filter = df[~df["身份证号"].isin(bad_ids)].copy()

# ========================
# 5. 随访阶段：用全部诊断列判断并发症
#    允许观测上出现 1 -> 5
# ========================
df_filter["heart_follow"] = df_filter[icd_cols].isin(heart_set).any(axis=1).astype(int)
df_filter["brain_follow"] = df_filter[icd_cols].isin(brain_set).any(axis=1).astype(int)
df_filter["kidney_follow"] = df_filter[icd_cols].isin(kidney_set).any(axis=1).astype(int)
```

```python
# ========================
# 6. 构造 df_states
#    - 基线固定从 state1 开始
#    - 后续按累计病史更新
#    - 首次进入 state5 后立即停止记录
#    - 末次状态不是5时才补行政截止点
# ========================
study_end = pd.Timestamp("2024-12-31")
all_results = []

groups = df_filter.groupby("身份证号", sort=False)

for pid, group in tqdm(groups, desc="Building df_states"):
    group = group.sort_values(["入院时间"]).copy()
    baseline_time = group.iloc[0]["入院时间"]

    heart_hist = 0
    brain_hist = 0
    kidney_hist = 0

    patient_rows = []

    for i, (_, row) in enumerate(group.iterrows()):
        # 首次住院固定视为 state1，不更新基线并发症历史
        if i > 0:
            if int(row["heart_follow"]) == 1:
                heart_hist = 1
            if int(row["brain_follow"]) == 1:
                brain_hist = 1
            if int(row["kidney_follow"]) == 1:
                kidney_hist = 1

        n_comp = heart_hist + brain_hist + kidney_hist

        if n_comp == 0:
            state = 1
        elif n_comp == 1:
            if heart_hist == 1:
                state = 2
            elif brain_hist == 1:
                state = 3
            elif kidney_hist == 1:
                state = 4
            else:
                raise ValueError(f"{pid} 的单并发症状态无法识别")
        else:
            state = 5

        time_days = (row["入院时间"] - baseline_time).days
        time_years = time_days / 365.25

        patient_rows.append({
            "身份证号": row["身份证号"],
            "payment": row["医疗付款方式"],
            "age": row["年龄"],
            "gender": row["性别"],
            "state": state,
            "time": row["入院时间"],
            "time_days": time_days,
            "time_years": time_years,
            "is_admin_censor": 0
        })

        # 一旦首次进入 state5，立即停止后续记录
        if state == 5:
            break

    # 若末次状态不是5，且末次时间早于研究截止，则补行政截止点
    last_state = patient_rows[-1]["state"]
    last_time = patient_rows[-1]["time"]

    if (last_state != 5) and (pd.notna(last_time)) and (last_time < study_end):
        time_days = (study_end - baseline_time).days
        time_years = time_days / 365.25

        patient_rows.append({
            "身份证号": patient_rows[-1]["身份证号"],
            "payment": patient_rows[-1]["payment"],
            "age": patient_rows[-1]["age"],
            "gender": patient_rows[-1]["gender"],
            "state": last_state,
            "time": study_end,
            "time_days": time_days,
            "time_years": time_years,
            "is_admin_censor": 1
        })

    all_results.extend(patient_rows)

df_states = pd.DataFrame(all_results)
```

```python
# ========================
# 7. 排序与检查
# ========================
df_states = df_states.sort_values(["身份证号", "time"]).reset_index(drop=True)
df_states["time_years_rounded"] = df_states["time_years"].round(4)

print("=" * 60)
print("状态分布：")
print(df_states["state"].value_counts().sort_index())

print("=" * 60)
print("行政补点数量：")
print(df_states["is_admin_censor"].sum())

print("=" * 60)
print("每位患者记录数分布（前10项）：")
print(df_states.groupby("身份证号").size().value_counts().sort_index().head(10))

# 相邻观测转移表
df_check = df_states.sort_values(["身份证号", "time"]).copy()
df_check["next_state"] = df_check.groupby("身份证号")["state"].shift(-1)

trans_table = (
    df_check.dropna(subset=["next_state"])
    .groupby(["state", "next_state"])
    .size()
    .reset_index(name="n")
    .sort_values(["state", "next_state"])
)

print("=" * 60)
print("相邻观测转移表：")
print(trans_table)
```
