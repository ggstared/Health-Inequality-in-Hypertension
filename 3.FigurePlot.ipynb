```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel(r"E:\vscode\Paper\Health  Inequalities in Hypertension\Revision1\File\Result\Part1\main\Part1-main.xlsx", sheet_name='TIE-line')
```

```python
# 总费用


plt.rcParams['font.family'] = 'Times New Roman'

x = [1, 2, 3, 4, 5, 6]

y_UEBMI = list(reversed(df['UEBMI'].tolist()))
yerr_UEBMI = list(reversed(df['UEBMI'] - df['UEBMI_lower']))

y_URRBMI = list(reversed(df['URRBMI'].tolist()))
yerr_URRBMI = list(reversed(df['URRBMI'] - df['URRBMI_lower']))

y_GS = list(reversed(df['GS'].tolist()))
yerr_GS = list(reversed(df['GS'] - df['GS_lower']))

y_COI = list(reversed(df['COI'].tolist()))
yerr_COI = list(reversed(df['COI'] - df['COI_lower']))
```

```python
plt.figure(figsize=(5, 3))

# 定义颜色
colors = {
    "UEBMI": "#BB4EA2", 
    "URRBMI": "#FBF165", 
    "GS": "#88C6E1", 
    "COI": "#8083B9"
}

# 绘图
plt.errorbar(x, y_UEBMI, yerr=yerr_UEBMI, fmt='o', color=colors["UEBMI"], capsize=3)
plt.plot(x, y_UEBMI, color=colors["UEBMI"], label="UEBMI")

plt.errorbar(x, y_URRBMI, yerr=yerr_URRBMI, fmt='o', color=colors["URRBMI"], capsize=3)
plt.plot(x, y_URRBMI, color=colors["URRBMI"], label="URRBMI")

plt.errorbar(x, y_GS, yerr=yerr_GS, fmt='o', color=colors["GS"], capsize=3)
plt.plot(x, y_GS, color=colors["GS"], label="GS")

plt.errorbar(x, y_COI, yerr=yerr_COI, fmt='o', color=colors["COI"], capsize=3)
plt.plot(x, y_COI, color=colors["COI"], label="COI")

# X轴年份
plt.xticks(x, ['1', '2', '3', '4', '5', '6'])

# 美化
plt.legend(frameon=False)                                     # 去掉图例边框
plt.tight_layout()

# 去掉上、右边框
ax = plt.gca()
for ytick in ax.get_yticks():
    ax.axhline(y=ytick, color="lightgray", linewidth=0.8, linestyle="--", zorder=0)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(left=False, bottom=False)  # 去掉刻度
ax.set_xlabel(' Observation window (years)')
ax.set_ylabel('Percent change in cost (%)')

ax.hlines(0, 0, 7, color="red", linewidth=0.8, linestyle="--", zorder=0, alpha=0.5)

# plt.show()
plt.tight_layout()
plt.savefig('File/Figure/Part1/main/TIE.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import matplotlib.patheffects as pe

_from = '4'
_to = '5'

# ======= 配置 =======
COL_EST = "r{}{}".format(_from, _to)   # 例：state1→state2 的占有概率
COL_L   = "r{}{}l".format(_from, _to)
COL_U   = "r{}{}u".format(_from, _to)
ymax = pd.to_numeric(df[COL_U].max())

# 颜色：仅 URBMI 可改，其它保持你的设定
colors = {
    "URRBMI": "#FBF165",  
    "UEBMI": "#BB4EA2",
    "COI": "#8083B9",
    "SP": "#B48581",
    "GS": "#88C6E1",
}
types_order = ["UEBMI", "URRBMI", "COI", "SP", "GS"]

# 全局风格
plt.rcParams.update({
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.grid": False,
    "grid.alpha": 0.25,
    "grid.linestyle": "--",
    "grid.linewidth": 0.6,
    "legend.frameon": False,
    "axes.labelsize": 11,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
})

fig, ax = plt.subplots(figsize=(5, 4), constrained_layout=True)

for typ in types_order:
    sub = (df.loc[df["type"] == typ, ["t", COL_EST, COL_L, COL_U]]
             .dropna()
             .sort_values("t"))
    if sub.empty:
        continue

    x  = sub["t"].to_numpy(float)
    y  = sub[COL_EST].to_numpy(float)
    yl = sub[COL_L].to_numpy(float)
    yu = sub[COL_U].to_numpy(float)

    # 先画线，再画置信带；URBMI 适当加粗并置于更高 zorder
    lw = 2.2 if typ == "URBMI" else 1.8
    z  = 3 if typ == "URBMI" else 2

    line, = ax.plot(
        x, y,
        label=typ,
        linewidth=lw,
        color=colors[typ],
        solid_capstyle="round",
        zorder=z,
        path_effects=[pe.Stroke(linewidth=lw+0.6, foreground="white", alpha=0.5), pe.Normal()]
    )
    ax.fill_between(x, yl, yu, color=colors[typ], alpha=0.18, zorder=1)

# 轴 & 刻度
ax.set_xlabel("Time (years)")
ax.set_ylabel("Pr[state = {} at time t | start in state {}]".format(_to, _from))
ax.set_xlim(left=0)  # 假设 t 从 0 开始
# ax.set_ylim(0, 1)    # 占有概率范围

# x 轴主刻度每 0.5，次刻度每 0.25
ax.xaxis.set_major_locator(MultipleLocator(0.5))
# ax.xaxis.set_minor_locator(MultipleLocator(0.25))

# y 轴可选设为百分比显示
# ax.yaxis.set_major_formatter(FuncFormatter(lambda v, pos: f"{v:.0%}"))

# 图例：两列、更紧凑
# ax.legend(
#     title="Payment type",
#     ncol=2,
#     handlelength=2.2,
#     columnspacing=0.8,
#     handletextpad=0.6,
#     fontsize=8.5,
#     # loc="upper left"
# )

# 轻微留白
ax.margins(x=0.02, y=0.05)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(left=False, bottom=False)  # 去掉刻度
for ytick in ax.get_yticks():
    if ytick < 0 or ytick > ymax:
        continue
    ax.axhline(y=ytick, color="lightgray", linewidth=0.8, linestyle="--", zorder=0)
plt.show()


# fig.savefig("File/Figure/Part2/main/pmatrix/state{}tostate{}.pdf".format(_from, _to), dpi=300, bbox_inches="tight")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# x 轴类别
labels = ["UEBMI", "URRBMI", "COI", "SP", "GS"]
x = np.arange(len(labels))

# ---- 数据堆叠：state1..state5（每个都是长度=5）----
data = np.vstack([y_state1, y_state2, y_state3, y_state4, y_state5]).astype(float)
# data = np.vstack([y_state2, y_state3, y_state4, y_state5]).astype(float)

# 可选健壮化：若不是严格和为1，自动按列归一化，避免堆叠>100%或<100%
colsum = data.sum(axis=0)
if not np.allclose(colsum, 1.0, atol=1e-6):
    data = data / colsum

# ---- 配色（按 state1..state5）----
colors = ["#EB6F5E", "#8BBCE8", "#C08EE3", "#8ED88E", "#E8B48B"]  # 去掉透明度，改由 alpha 控制
# colors = ["#8BBCE8", "#C08EE3", "#8ED88E", "#E8B48B"]  # 去掉透明度，改由 alpha 控制
edgecolor = "white"

fig, ax = plt.subplots(figsize=(5, 4), constrained_layout=True)

bottom = np.zeros(len(labels))
bars = []
for k, (yk, c) in enumerate(zip(data, colors), start=1):
    b = ax.bar(
        x, yk, bottom=bottom, width=0.62,
        color=c, alpha=0.95,
        edgecolor=edgecolor, linewidth=0.8,
        label=f"state{k}", zorder=2
    )
    bars.append(b)
    bottom += yk

# ---- 轴 & 样式 ----
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=0, fontsize=7.5)
ax.set_ylim(0.9, 1)
ax.yaxis.set_major_formatter(PercentFormatter(1.0))   # 百分比
ax.set_ylabel("State distribution at t = 5 years (%)")  # 或者：ax.set_ylabel("截至 5 年的状态分布（%）")

# 网格只画 y 轴，淡一点
ax.grid(axis='y', alpha=0.25, linestyle='--', linewidth=0.7, zorder=1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# ---- 段内数值标签：仅在占比≥4%时标注 ----
for b, yk in zip(bars, data):
    ax.bar_label(
        b,
        # labels=[f"{v*100:.0f}%" if v >= 0.04 else "" for v in yk],
        labels=[f"{v*100:.2f}%" for v in yk],
        label_type='center', fontsize=7.5, color="black"
    )

# ---- 图例：倒序（state5在上），更紧凑 ----
handles, labels_legend = ax.get_legend_handles_labels()
# ax.legend(
#     handles[::-1], labels_legend[::-1],
#     title="State", ncol=1, frameon=False,
#     bbox_to_anchor=(1.01, 1), loc="upper left", borderaxespad=0.0
# )

ax.set_title("Distribution of states at t = 5 years by payment type")

plt.show()

# 可选导出
fig.savefig("File/Figure/Part2/main/totlos/totlos.pdf", dpi=300, bbox_inches="tight")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# x 轴类别
labels = ["UEBMI", "URRBMI", "SP"]
x = np.arange(len(labels))

# ---- 数据堆叠：state1..state5（每个都是长度=5）----
data = np.vstack([y_state1, y_state2, y_state3, y_state4, y_state5]).astype(float)

# 可选健壮化：若不是严格和为1，自动按列归一化，避免堆叠>100%或<100%
colsum = data.sum(axis=0)
if not np.allclose(colsum, 1.0, atol=1e-6):
    data = data / colsum

# ---- 配色（按 state1..state5）----
colors = ["#EB6F5E", "#8BBCE8", "#C08EE3", "#8ED88E", "#E8B48B"]  # 去掉透明度，改由 alpha 控制
edgecolor = "white"

fig, ax = plt.subplots(figsize=(5, 4), constrained_layout=True)

bottom = np.zeros(len(labels))
bars = []
for k, (yk, c) in enumerate(zip(data, colors), start=1):
    b = ax.bar(
        x, yk, bottom=bottom, width=0.62,
        color=c, alpha=0.95,
        edgecolor=edgecolor, linewidth=0.8,
        label=f"state{k}", zorder=2
    )
    bars.append(b)
    bottom += yk

# ---- 轴 & 样式 ----
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=0, fontsize=7.5)
ax.set_ylim(0.9, 1)
ax.yaxis.set_major_formatter(PercentFormatter(1.0))   # 百分比
ax.set_ylabel("State distribution at t = 5 years (%)")  # 或者：ax.set_ylabel("截至 5 年的状态分布（%）")

# 网格只画 y 轴，淡一点
ax.grid(axis='y', alpha=0.25, linestyle='--', linewidth=0.7, zorder=1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# ---- 段内数值标签：仅在占比≥4%时标注 ----
for b, yk in zip(bars, data):
    ax.bar_label(
        b,
        # labels=[f"{v*100:.0f}%" if v >= 0.04 else "" for v in yk],
        labels=[f"{v*100:.2f}%" for v in yk],
        label_type='center', fontsize=7.5, color="black"
    )

# ---- 图例：倒序（state5在上），更紧凑 ----
handles, labels_legend = ax.get_legend_handles_labels()
# ax.legend(
#     handles[::-1], labels_legend[::-1],
#     title="State", ncol=1, frameon=False,
#     bbox_to_anchor=(1.01, 1), loc="upper left", borderaxespad=0.0
# )

ax.set_title("Distribution of states at t = 5 years by payment type")

plt.show()

# 可选导出
fig.savefig("File/Figure/Part2/External-validation/totlos/totlos.pdf", dpi=300, bbox_inches="tight")
```
